# Overview of credit card fraud and its impact on individuals and businesses. 
Credit card fraud is a prevalent form of financial crime that significantly impacts both individuals and businesses. This illicit activity involves unauthorized use of credit or debit card information to make fraudulent transactions, posing substantial risks to financial security. For individuals, credit card fraud can result in financial loss, identity theft, and damage to credit scores. In addition to the direct financial consequences, victims often face the time-consuming and stressful process of resolving fraudulent charges and restoring their financial reputation.

Businesses, on the other hand, bear the brunt of credit card fraud through financial losses, increased operational costs related to fraud detection and prevention, and potential damage to their reputation. The impact extends beyond monetary losses, as businesses must invest in advanced security measures to safeguard customer data and maintain trust. Additionally, regulatory consequences and legal liabilities may arise if businesses fail to adequately protect customer information.

Overall, credit card fraud poses a pervasive threat, necessitating continuous advancements in fraud detection technologies and collaborative efforts between financial institutions, businesses, and individuals to mitigate its impact on both personal finances and the broader economy.

Reputation Management: Instances of fraud can tarnish the reputation of businesses and financial institutions. Developing and implementing effective fraud detection methods helps protect the reputation of organizations by demonstrating a commitment to security and ethical business practices.

Global Economic Impact: Fraudulent activities, especially on a large scale, can have broader economic implications. By minimizing fraud, effective detection methods contribute to the stability and trustworthiness of financial systems, positively influencing the overall economy.

Data Protection: Fraud detection methods often involve robust data protection measures. Implementing these methods not only helps prevent financial fraud but also ensures the security and privacy of sensitive customer information, aligning with evolving data protection regulations.

In summary, the significance of developing effective fraud detection methods extends beyond individual and business protection; it encompasses trust-building, regulatory compliance, operational efficiency, and the overall health of the global economy.

# Summarize various approaches, algorithms, and machine learning models used in previous research. 
Supervised Learning Algorithms: Researchers have extensively applied supervised learning algorithms, including logistic regression, decision trees, support vector machines, and ensemble methods such as random forests and gradient boosting. These models leverage labeled datasets to learn patterns indicative of fraudulent transactions.

Unsupervised Learning and Anomaly Detection: Unsupervised learning techniques, particularly anomaly detection methods, have gained attention. Approaches like k-means clustering, isolation forests, and autoencoders aim to identify abnormal patterns in transaction data without requiring labeled examples of fraud.

Ensemble Methods: Combining multiple models into ensemble methods has been a common strategy to enhance the robustness and accuracy of fraud detection systems. Ensemble methods, like stacking or bagging, leverage the strengths of different algorithms.

# Strengths and limitations of different methodologies.
Supervised Learning Algorithms:

Strengths: Effective with Labeled Data: Supervised learning algorithms perform well when provided with labeled datasets, making them suitable for fraud detection tasks with historical fraud/non-fraud labels.
Interpretability: Many supervised models, such as decision trees and logistic regression, offer interpretability, allowing analysts to understand the factors influencing predictions.

Limitations:Dependence on Labeled Data: The reliance on labeled data can be a limitation, as acquiring accurately labeled instances of fraud may be challenging.
Difficulty in Capturing Novelty: Supervised models may struggle to detect novel or previously unseen forms of fraud.

Unsupervised Learning and Anomaly Detection:

Strengths: Detection of Novel Patterns: Anomaly detection methods excel at identifying unusual patterns without requiring labeled data, making them effective in recognizing novel fraud instances.
No Prior Assumptions: Unsupervised methods do not assume prior knowledge about fraud patterns, allowing them to adapt to changing fraud tactics.

Limitations: False Positives: Anomaly detection may generate false positives, as legitimate but uncommon transactions can be flagged as anomalies.

Scalability Issues: Some unsupervised methods may face challenges in scaling to large datasets or real-time processing.

**Information about the Data**

The dataset contains transactions made by credit cards in September 2013 by European cardholders. This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

It contains only numerical input variables which are the result of a PCA transformation. Unfortunately, due to confidentiality issues, we cannot provide the original features and more background information about the data. Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

Given the class imbalance ratio, we recommend measuring the accuracy using the Area Under the Precision-Recall Curve (AUPRC). Confusion matrix accuracy is not meaningful for unbalanced classification.

DATA PREPROCESSING

```python
# Distribution of Fraud and Normal Transactions

px.pie(df, names="Class", title="Distribution of Fraud and Normal Transactions:(Normal:0 | Fraud:1)")
```

```python
# making Hour feature from Time feature
df["Hour"] = df["Time"].apply(lambda x: np.ceil(float(x)/3600) % 24)
df["Hour"] = df["Hour"].astype("int")

# time vs fraud

# Transaction count with non frauds over time(Hour)
count_0 = df[df['Class']==0].groupby('Hour').count()["Class"]

# Transaction count with frauds over time(Hour)
count_1 = df[df['Class']==1].groupby('Hour').count()["Class"]

# Concatenate
counts_df = pd.concat([count_0, count_1], axis=1, keys=["Class 0", "Class 1"]).fillna(0)
df_counts = counts_df.reset_index()


# Subplots
fig = make_subplots(
    rows=1, cols=2,

    subplot_titles=("Non Frauds","Frauds"))
fig.add_trace(go.Scatter(x=df_counts["Hour"], y=df_counts["Class 0"], name="Transactions", marker = {'color' : "#1E90FF"}),
                 row=1, col=1)

fig.add_trace(go.Scatter(x=df_counts["Hour"], y=df_counts["Class 1"], name="Transactions", marker = {'color' : "#FF4500"}),
                 row=1, col=2)

fig['layout']['xaxis']['title']='Hour'
fig['layout']['yaxis']['title']='Transaction'
fig['layout']['xaxis2']['title']='Hour'
fig['layout']['yaxis2']['title']='Transaction'

fig.show()
```

## MODEL BUILDING

```python
X=df.drop(["Class"],axis=1).values
y=df["Class"].values

# X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=7,stratify=y)
# X.shape
nm=NearMiss()
X_res,y_res=nm.fit_resample(X,y)
X_res.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.3,random_state=7,stratify=y_res)
```

```python
# standardization of data
sc=StandardScaler()

X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
def model_val(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print(model)

    print(f"Test Data Accuracy: {accuracy_score(y_test,y_pred)}")
    print("Model Test F1 Score is: ", f1_score(y_test, y_pred))
    print("Classification_report: \n",classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    # Print confusion matrix
    print("Confusion Matrix:")
    print(cm)
```

```python
lr = LogisticRegression()
model_val(lr,X_train,X_test,y_train,y_test)
```

```python
svc = SVC()
model_val(svc,X_train,X_test,y_train,y_test)
```

```python
knn=KNeighborsClassifier()
model_val(knn,X_train,X_test,y_train,y_test)
```

```python
dtc=DecisionTreeClassifier()
model_val(dtc,X_train,X_test,y_train,y_test)
```

```python
nb=GaussianNB()
model_val(nb,X_train,X_test,y_train,y_test)
```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
y_pred=dt_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print('Decision Tree Classifier')
print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
print("\nConfusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Calculate ROC curve values
y_prob = dt_classifier.predict_proba(X_test)[:, 1]  # Probability of positive class
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)


feature_importances = dt_classifier.feature_importances_
feature_names = df.columns  # Assuming X_train is a DataFrame
```

```python
plt.figure(figsize=(18, 6))

# Confusion Matrix Heatmap
plt.subplot(1, 3, 1)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")

# AUC-ROC Curve
plt.subplot(1, 3, 2)
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

# Feature Importances Diagram
plt.subplot(1, 3, 3)
sorted_idx = feature_importances.argsort()
plt.barh(range(len(sorted_idx)), feature_importances[sorted_idx], align="center")
plt.yticks(range(len(sorted_idx)), [feature_names[i] for i in sorted_idx])
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Feature Importances")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
